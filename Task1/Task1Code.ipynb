```python
# Importing the pandas libraries
import pandas as pd
```

```python
# Loading the dataset and viewing the data
data=pd.read_csv("KaggleV2-May-2016.csv")
data.head()
```

```python
#Identifying missing values
data.isnull().sum()
```

```python
# Dropping duplicates
data.drop_duplicates() 
```

```python
data.columns
```

```python
#Changing the column names
data.columns=data.columns.str.strip().str.replace("Handcap","Handicap").str.replace("-","_").str.replace("Hipertension","Hypertension")
```

```python
# Checking the datatypes
print(data["Age"].dtype)
print(data["ScheduledDay"].dtype)
print(data["AppointmentDay"].dtype)
```

```python
#Changing the columns to type date time
data["ScheduledDay"]=pd.to_datetime(data["ScheduledDay"])
data["AppointmentDay"]=pd.to_datetime(data["AppointmentDay"])
```

```python
print(data.dtypes)
```

```python
# Standardize text values for gender
data['Gender'] = data['Gender'].str.upper().str.strip()
```

```python
data.head()
```

```python
# Saving the cleaned dataset
data.to_csv("Cleaned_KaggleV2_May_2016.csv", index=False)
```
